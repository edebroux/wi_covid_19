````python
# Load the dataset from the datasets folder
state_data = pd.read_csv("./wisconsin_data/10012020/wisconsin_covid_19_data.csv", encoding = "utf-8")
state_data = state_data.loc[state_data["NAME"] == "WI"].sort_values(by = "POSITIVE").reset_index()
state_data = state_data.iloc[:, 2:]
# View the first five rows of the statewide dataset
state_data.head()
````

We want to plot the dataset information. We want to do a few things here:

- Plot the cummulative tests, cummulative positive and negative case counts, and cummulative death counts.
- Plot the 7-day moving average for these same statistics, as most coronavirus cases have a lag between when the disease is contracted and when they test positive
- Plot the daily positive COVID-19 test rate percentage.

````python
def plt_covid_cum(dates, num_pos, num_neg, num_dth):
    plt.figure(figsize = (12, 8))
    ax = plt.subplot()
    # Plot the number of tests, positive tests, and deaths
    plt.plot(range(len(dates)), num_pos + num_neg, color = "blue", linewidth = 2,
            label = "Number of Tests")
    plt.plot(range(len(dates)), num_pos, color = "red", linewidth = 2,
            label = "Number of Positive Tests")
    #plt.plot(range(len(dates)), num_neg, color = "green", linewidth = 2,
    #        label = Number of Negative Tests")
    plt.plot(range(len(dates)), num_dth, color = "black", linewidth = 2,
            label = "Number of Deaths from COVID-19")
    plt.title("Wisconsin Cummulative COVID-19 Counts")
    # Annotate certain key dates in the Wisconsin COVID-19 timeline
    # First date Wisconsin DHS creates a dashboard tracking cases
    plt.annotate("1",
                 xy = (0, num_pos[0]),
                 xytext = (0, num_pos[0] + 200000),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # All gatherings of more than 10 people banned
    plt.annotate("2",
                 xy = (2, num_pos[2]),
                 xytext = (2, num_pos[2] + 200000),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # All state schools close, March 18th
    plt.annotate("3",
                 xy = (5, num_pos[3]),
                 xytext = (5, num_pos[3] + 200000),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # All non-essential businesses are ordered to close, March 24th
    plt.annotate("4",
                xy = (8, num_pos[8]),
                xytext = (8, num_pos[8] + 200000),
                arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Safer-at-home order extended to May 26th, April 16th
    plt.annotate("5",
                xy = (31, num_pos[31]),
                xytext = (31, num_pos[31] + 200000),
                arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Anti-lockdown protests at State Capitol, April 24th
    plt.annotate("6",
                 xy = (39, num_pos[39]),
                 xytext = (39, num_pos[39] + 200000),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Safer-at-home order struck down by Wisconsin Supreme Court, May 13th
    plt.annotate("7",
                 xy = (57, num_pos[57]),
                 xytext = (57, num_pos[57] + 200000),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # George Floyd is killed by MPLS Police, May 25th
    plt.annotate("8",
                xy = (70, num_pos[70]),
                xytext = (70, num_pos[70] + 200000),
                arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # July 4th
    plt.annotate("9",
                 xy = (110, num_pos[110]),
                 xytext = (110, num_pos[110] + 200000),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Governor Evers declares public health emergency requiring masks to be worn
    # in enclosed spaces outside private residences, July 30th
    plt.annotate("10",
                 xy = (136, num_pos[136]),
                 xytext = (136, num_pos[136] + 200000),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Labor Day, September 7th. Public schools to start class or are entering their second week
    plt.annotate("11",
                xy = (175, num_pos[175]),
                xytext = (175, num_pos[175] + 200000),
                arrowprops = dict(facecolor = 'black', shrink = 0.05))
    
    plt.xlabel("Date")
    plt.ylabel("Number of Cases")
    plt.legend()
    ax.set_xticks(range(len(dates)))
    ax.set_xticklabels(dates)
    # TO-DO: add xticks for every week
    for idx, lbl in enumerate(ax.get_xticklabels()):
        if idx % 14 != 0:
            lbl.set_visible(False)
    # Make xticks more easily visible
    plt.xticks(rotation = 300)
    
def plt_positive_percentage(dates, new_pos, new_tests):
    plt.figure(figsize = (12, 8))
    percent_pos = (np.array(new_pos).astype(float) / (np.array(new_tests).astype(float))) * 100.
    percent_pos[0] = (32. / (32. + 313.)) * 100.
    # Creating a 7-day moving average calculation to write over the graphic
    seven_day_pos = []
    seven_day_tst = []
    # Basic loop to calculate the 7-day moving average
    for i in range(len(new_tests) - 6):
        day_pos = np.sum(new_pos[i:(i+7)]) / 7.
        day_tst = np.sum(new_tests[i:(i+7)]) / 7.
        seven_day_pos.append(day_pos)
        seven_day_tst.append(day_tst)
    # Print the seven day moving average of positive cases
    #print(seven_day_pos, seven_day_tst)
    # Calculating the 7-day moving average positive test percentage rate
    seven_ma_percent_pos = (np.array(seven_day_pos).astype(float) / np.array(seven_day_tst).astype(float)) * 100.
    ax = plt.subplot()
    plt.plot(range(len(dates)), percent_pos, color = "red",
             linewidth = 2, label = "Daily Percent of Positive Tests")
    plt.plot(range(6, len(dates)), seven_ma_percent_pos, color = "darkgreen",
             linewidth = 2, label = "7-day Moving Average Percent of Positive Tests")
    plt.ylim(-20, 35)
    # Annotate certain key dates in the Wisconsin COVID-19 timeline
    # First date Wisconsin DHS creates a dashboard tracking cases
    plt.annotate("1",
                 xy = (0, percent_pos[0]),
                 xytext = (0, percent_pos[0] + 10),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # All gatherings of more than 10 people banned
    plt.annotate("2",
                 xy = (2, percent_pos[2]),
                 xytext = (2, percent_pos[2] + 10),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # All state schools close, March 18th
    plt.annotate("3",
                 xy = (5, percent_pos[3]),
                 xytext = (5, percent_pos[3] + 10),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # All non-essential businesses are ordered to close, March 24th
    plt.annotate("4",
                xy = (8, percent_pos[4]),
                xytext = (8, percent_pos[4] + 10),
                arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # State deletes all non-Wisconsin resident tests from their dashboard
    plt.annotate("5",
                xy = (15, percent_pos[15]),
                xytext = (15, percent_pos[15] + 10),
                arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Safer-at-home order extended to May 26th, April 16th
    plt.annotate("6",
                xy = (31, percent_pos[31]),
                xytext = (31, percent_pos[31] + 10),
                arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Anti-lockdown protests at State Capitol, April 24th
    plt.annotate("7",
                 xy = (39, percent_pos[39]),
                 xytext = (39, percent_pos[39] + 10),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Safer-at-home order struck down by Wisconsin Supreme Court, May 13th
    plt.annotate("8",
                 xy = (57, percent_pos[57]),
                 xytext = (57, percent_pos[57] + 10),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # George Floyd is killed by MPLS Police, May 25th
    plt.annotate("9",
                xy = (70, percent_pos[70]),
                xytext = (70, percent_pos[70] + 10),
                arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # July 4th
    plt.annotate("10",
                 xy = (110, percent_pos[110]),
                 xytext = (110, percent_pos[110] + 10),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Governor Evers declares public health emergency requiring masks to be worn
    # in enclosed spaces outside private residences, July 30th
    plt.annotate("11",
                 xy = (136, percent_pos[136]),
                 xytext = (136, percent_pos[136] + 10),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Labor Day, September 7th. Public schools to start class or are entering their second week
    plt.annotate("12",
                xy = (175, percent_pos[175]),
                xytext = (175, percent_pos[175] + 10),
                arrowprops = dict(facecolor = 'black', shrink = 0.05))
    plt.title("Wisconsin COVID-19 Percentages")
    plt.xlabel("Date")
    plt.ylabel("Percentage")
    plt.legend()
    ax.set_xticks(range(len(dates)))
    ax.set_xticklabels(dates)
    # TO-DO: add xticks for every week
    for idx, lbl in enumerate(ax.get_xticklabels()):
        if idx % 14 != 0:
            lbl.set_visible(False)
    plt.xticks(rotation = 300)

# THIS NEEDS A REWRITE
def plt_seven_day(dates, new_pos, new_tests, new_dth):
    seven_day_pos = []
    seven_day_tst = []
    seven_day_dth = []
    
    # Basic loop to calculate the 7-day moving average
    for i in range(len(new_tests) - 6):
        day_pos = np.sum(new_pos[i:(i+7)]) / 7.
        day_tst = np.sum(new_tests[i:(i+7)]) / 7.
        day_dth = np.sum(new_dth[i:(i+7)]) / 7.
        seven_day_pos.append(day_pos)
        seven_day_tst.append(day_tst)
        seven_day_dth.append(day_dth)
        
    plt.figure(figsize = (12, 8))
    ax = plt.subplot()
    plt.plot(range(len(seven_day_tst)), pd.Series(seven_day_tst), color = "blue", linewidth = 2,
            label = "Average Number of Tests")
    plt.plot(range(len(seven_day_pos)), seven_day_pos, color = "red", linewidth = 2,
            label = "Average Number of Positive Tests")
    plt.plot(range(len(seven_day_dth)), seven_day_dth, color = "black", linewidth = 2,
            label = "Average Number of Deaths")
    plt.title("Wisconsin COVID-19 7-day Moving Averages")
    plt.xlabel("Dates")
    plt.ylabel("Average Number of Cases")
    plt.legend()
    # Annotate certain key dates in the Wisconsin COVID-19 timeline
    # First date Wisconsin DHS creates a dashboard tracking cases
    plt.annotate("1",
                 xy = (0, seven_day_pos[0]),
                 xytext = (0, seven_day_pos[0] + 1000),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # All gatherings of more than 10 people banned
    plt.annotate("2",
                 xy = (2, seven_day_pos[2]),
                 xytext = (2, seven_day_pos[2] + 1000),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # All state schools close, March 18th
    plt.annotate("3",
                 xy = (3, seven_day_pos[3]),
                 xytext = (3, seven_day_pos[3] + 1000),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # All non-essential businesses are ordered to close, March 24th
    plt.annotate("4",
                 xy = (8, seven_day_pos[4]),
                 xytext = (8, seven_day_pos[4] + 1000),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Tests from non-Wisconsin residents removed from dataset
    plt.annotate("5",
                 xy = (15, seven_day_pos[15]),
                 xytext = (15, seven_day_pos[15] + 1000),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Safer-at-home order extended to May 26th, April 16th
    plt.annotate("6",
                 xy = (31, seven_day_pos[31]),
                 xytext = (31, seven_day_pos[31] + 1000),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Anti-lockdown protests at State Capitol, April 24th
    plt.annotate("7",
                  xy = (39, seven_day_pos[39]),
                  xytext = (39, seven_day_pos[39] + 1000),
                  arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Safer-at-home order struck down by Wisconsin Supreme Court, May 13th
    plt.annotate("8",
                  xy = (57, seven_day_pos[57]),
                  xytext = (57, seven_day_pos[57] + 1000),
                  arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # George Floyd is killed by MPLS Police, May 25th
    plt.annotate("9",
                 xy = (70, seven_day_pos[70]),
                 xytext = (70, seven_day_pos[70] + 1000),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # July 4th
    plt.annotate("10",
                  xy = (110, seven_day_pos[110]),
                  xytext = (110, seven_day_pos[110] + 1000),
                  arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Governor Evers declares public health emergency requiring masks to be worn
    # in enclosed spaces outside private residences, July 30th
    plt.annotate("11",
                  xy = (136, seven_day_pos[136]),
                  xytext = (136, seven_day_pos[136] + 1000),
                  arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Labor Day, September 7th. Public schools to start class or are entering their second week
    plt.annotate("12",
                 xy = (175, seven_day_pos[175]),
                 xytext = (175, seven_day_pos[175] + 1000),
                 arrowprops = dict(facecolor = 'black', shrink = 0.05))
    ax.set_xticks(range(len(dates)))
    ax.set_xticklabels(dates)
    for idx, lbl in enumerate(ax.get_xticklabels()):
        if idx % 14 != 0:
            lbl.set_visible(False)
    plt.xticks(rotation = 300)
````

````python
def daily_infection_rates_ages(total, ages_0_9, ages_10_19, ages_20_29, ages_30_39, ages_40_49, ages_50_59,
                 ages_60_69, ages_70_79, ages_80_89, ages_90, dates, title1, title2, ylabel, arrow_adj = 0):
    # Calculate and plot the daily increase/decrease in infections, hospitalizations, ICU patients
    daily_0_9   = pd.Series(ages_0_9).diff()
    daily_10_19 = pd.Series(ages_10_19).diff()
    daily_20_29 = pd.Series(ages_20_29).diff()
    daily_30_39 = pd.Series(ages_30_39).diff()
    daily_40_49 = pd.Series(ages_40_49).diff()
    daily_50_59 = pd.Series(ages_50_59).diff()
    daily_60_69 = pd.Series(ages_60_69).diff()
    daily_70_79 = pd.Series(ages_70_79).diff()
    daily_80_89 = pd.Series(ages_80_89).diff()
    daily_90    = pd.Series(ages_90).diff()
    daily_0_9[0] = 0; daily_10_19[0] = 0; daily_20_29[0] = 0; daily_30_39[0] = 0; daily_40_49[0] = 0
    daily_50_59[0] = 0; daily_60_69[0] = 0; daily_70_79[0] = 0; daily_80_89[0] = 0; daily_90[0] = 0
    # Plot these demographic pieces of information
    daily_nir = {
        'a09': np.zeros(len(ages_0_9)),
        'a1019': np.zeros(len(ages_10_19)),
        'a2029': np.zeros(len(ages_20_29)),
        'a3039': np.zeros(len(ages_30_39)),
        'a4049': np.zeros(len(ages_40_49)),
        'a5059': np.zeros(len(ages_50_59)),
        'a6069': np.zeros(len(ages_60_69)),
        'a7079': np.zeros(len(ages_70_79)),
        'a8089': np.zeros(len(ages_80_89)),
        'a90': np.zeros(len(ages_90)),
        'dates': dates
    }
    seven_day = pd.DataFrame(daily_nir)
    for i in range(len(daily_0_9[:6]), len(daily_0_9)):
        seven_day.a09[i] = (np.sum(np.array(daily_0_9[i:(i+7)]).astype(float)) / 7.)
        seven_day.a1019[i] = (np.sum(np.array(daily_10_19[i:(i+7)]).astype(float)) / 7.)
        seven_day.a2029[i] = (np.sum(np.array(daily_20_29[i:(i+7)]).astype(float)) / 7.)
        seven_day.a3039[i] = (np.sum(np.array(daily_30_39[i:(i+7)]).astype(float)) / 7.)
        seven_day.a4049[i] = (np.sum(np.array(daily_40_49[i:(i+7)]).astype(float)) / 7.)
        seven_day.a5059[i] = (np.sum(np.array(daily_50_59[i:(i+7)]).astype(float)) / 7.)
        seven_day.a6069[i] = (np.sum(np.array(daily_60_69[i:(i+7)]).astype(float)) / 7.)
        seven_day.a7079[i] = (np.sum(np.array(daily_70_79[i:(i+7)]).astype(float)) / 7.)
        seven_day.a8089[i] = (np.sum(np.array(daily_80_89[i:(i+7)]).astype(float)) / 7.)
        seven_day.a90[i] = (np.sum(np.array(daily_90[i:(i+7)]).astype(float)) / 7.)
    # Plot these seven-day infection rates and the moving averages
    colors = ["tab:blue", "tab:orange", "tab:green", "tab:red", "tab:purple",
             "tab:brown", "tab:pink", "tab:gray", "tab:olive", "tab:cyan"]
    
    plt.figure(figsize = (12, 16))
    ax1 = plt.subplot(211)
    ax1.bar(range(len(daily_0_9)), daily_0_9, width = 0.5, color = colors[0], label = "Ages 0-9")
    ax1.bar(range(len(daily_10_19)), daily_10_19, width = 0.5, bottom = daily_0_9, color = colors[1], label = "Ages 10-19")
    ax1.bar(range(len(daily_20_29)), daily_20_29, width = 0.5, bottom = daily_10_19, color = colors[2], label = "Ages 20-29")
    ax1.bar(range(len(daily_30_39)), daily_30_39, width = 0.5, bottom = daily_20_29, color = colors[3], label = "Ages 30-39")
    ax1.bar(range(len(daily_40_49)), daily_40_49, width = 0.5, bottom = daily_30_39, color = colors[4], label = "Ages 40-49")
    ax1.bar(range(len(daily_50_59)), daily_50_59, width = 0.5, bottom = daily_40_49, color = colors[5], label = "Ages 50-59")
    ax1.bar(range(len(daily_60_69)), daily_60_69, width = 0.5, bottom = daily_50_59, color = colors[6], label = "Ages 60-69")
    ax1.bar(range(len(daily_70_79)), daily_70_79, width = 0.5, bottom = daily_60_69, color = colors[7], label = "Ages 70-79")
    ax1.bar(range(len(daily_80_89)), daily_80_89, width = 0.5, bottom = daily_70_79, color = colors[8], label = "Ages 80-89")
    ax1.bar(range(len(daily_90)), daily_90, width = 0.5, bottom = daily_80_89, color = colors[9], label = "Ages 90 and Over")
    ax1.legend()
    ax1.set_xlabel("Date")
    ax1.set_ylabel(ylabel)
    ax1.set_title(title1)
    # Annotate certain key dates in the Wisconsin COVID-19 timeline
    # First date Wisconsin DHS creates a dashboard tracking cases
    #ax1.annotate("1",
    #             xy = (0, total[0]),
    #             xytext = (0, total[0] + arrow_adj),
    #             arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # All gatherings of more than 10 people banned
    #ax1.annotate("2",
    #             xy = (2, total[2]),
    #             xytext = (2, total[2] + arrow_adj),
    #             arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # All state schools close, March 18th
    #ax1.annotate("3",
    #             xy = (5, total[3]),
    #             xytext = (5, total[3] + arrow_adj),
    #             arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # All non-essential businesses are ordered to close, March 24th
    #ax1.annotate("4",
    #            xy = (8, total[8]),
    #            xytext = (8, total[8] + arrow_adj),
    #            arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Safer-at-home order extended to May 26th, April 16th
    #ax1.annotate("5",
    #            xy = (31, total[31]),
    #            xytext = (31, total[31] + arrow_adj),
    #            arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Anti-lockdown protests at State Capitol, April 24th
    #ax1.annotate("6",
    #             xy = (39, total[39]),
    #             xytext = (39, total[39] + arrow_adj),
    #             arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Safer-at-home order struck down by Wisconsin Supreme Court, May 13th
    #ax1.annotate("7",
    #             xy = (57, total[57]),
    #             xytext = (57, total[57] + arrow_adj),
    #             arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # George Floyd is killed by MPLS Police, May 25th
    #ax1.annotate("8",
    #            xy = (70, total[70]),
    #            xytext = (70, total[70] + arrow_adj),
    #            arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # July 4th
    #ax1.annotate("9",
    #             xy = (110, total[110]),
    #             xytext = (110, total[110] + arrow_adj),
    #             arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Governor Evers declares public health emergency requiring masks to be worn
    # in enclosed spaces outside private residences, July 30th
    #ax1.annotate("10",
    #             xy = (136, total[136]),
    #             xytext = (136, total[136] + arrow_adj),
    #             arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Labor Day, September 7th. Public schools to start class or are entering their second week
    #ax1.annotate("11",
    #            xy = (175, total[175]),
    #            xytext = (175, total[175] + arrow_adj),
    #            arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # TO-DO: add xticks for every week
    ax1.set_xticks(range(len(daily_0_9)))
    ax1.set_xticklabels(dates, rotation = 300)
    # TO-DO: add xticks for every week
    for idx, lbl in enumerate(ax1.get_xticklabels()):
        if idx % 14 != 0:
            lbl.set_visible(False)
            
    ax2 = plt.subplot(212)
    ax2.bar(range(len(seven_day.a09)), seven_day.a09, width = 0.5,
            color = colors[0], label = "Ages 0-9")
    ax2.bar(range(len(seven_day.a1019)), seven_day.a1019, width = 0.5,
            bottom = seven_day.a09, color = colors[1], label = "Ages 10-19")
    ax2.bar(range(len(seven_day.a2029)), seven_day.a2029, width = 0.5,
            bottom = seven_day.a1019, color = colors[2], label = "Ages 20-29")
    ax2.bar(range(len(seven_day.a3039)), seven_day.a3039, width = 0.5,
            bottom = seven_day.a2029, color = colors[3], label = "Ages 30-39")
    ax2.bar(range(len(seven_day.a4049)), seven_day.a4049, width = 0.5,
            bottom = seven_day.a3039, color = colors[4], label = "Ages 40-49")
    ax2.bar(range(len(seven_day.a5059)), seven_day.a5059, width = 0.5,
            bottom = seven_day.a4049, color = colors[5], label = "Ages 50-59")
    ax2.bar(range(len(seven_day.a6069)), seven_day.a6069, width = 0.5,
            bottom = seven_day.a5059, color = colors[6], label = "Ages 60-69")
    ax2.bar(range(len(seven_day.a7079)), seven_day.a7079, width = 0.5,
            bottom = seven_day.a6069, color = colors[7], label = "Ages 70-79")
    ax2.bar(range(len(seven_day.a8089)), seven_day.a8089, width = 0.5,
            bottom = seven_day.a7079, color = colors[8], label = "Ages 80-89")
    ax2.bar(range(len(seven_day.a90)), seven_day.a90, width = 0.5,
            bottom = seven_day.a8089, color = colors[9], label = "Ages 90 and Over")
    ax2.legend()
    ax2.set_xlabel("Date")
    ax2.set_ylabel(ylabel)
    ax2.set_title(title2)
    # Annotate certain key dates in the Wisconsin COVID-19 timeline
    # First date Wisconsin DHS creates a dashboard tracking cases
    #ax2.annotate("1",
    #             xy = (0, total[0]),
    #             xytext = (0, total[0] + arrow_adj),
    #             arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # All gatherings of more than 10 people banned
    #ax2.annotate("2",
    #             xy = (2, total[2]),
    #             xytext = (2, total[2] + arrow_adj),
    #             arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # All state schools close, March 18th
    #ax2.annotate("3",
    #             xy = (5, total[3]),
    #             xytext = (5, total[3] + arrow_adj),
    #             arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # All non-essential businesses are ordered to close, March 24th
    #ax2.annotate("4",
    #            xy = (8, total[8]),
    #            xytext = (8, total[8] + arrow_adj),
    #            arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Safer-at-home order extended to May 26th, April 16th
    #ax2.annotate("5",
    #            xy = (31, total[31]),
    #            xytext = (31, total[31] + arrow_adj),
    #            arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Anti-lockdown protests at State Capitol, April 24th
    #ax2.annotate("6",
    #             xy = (39, total[39]),
    #             xytext = (39, total[39] + arrow_adj),
    #             arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Safer-at-home order struck down by Wisconsin Supreme Court, May 13th
    #ax2.annotate("7",
    #             xy = (57, total[57]),
    #             xytext = (57, total[57] + arrow_adj),
    #             arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # George Floyd is killed by MPLS Police, May 25th
    #ax2.annotate("8",
    #            xy = (70, total[70]),
    #            xytext = (70, total[70] + arrow_adj),
    #            arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # July 4th
    #ax2.annotate("9",
    #             xy = (110, total[110]),
    #             xytext = (110, total[110] + arrow_adj),
    #             arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Governor Evers declares public health emergency requiring masks to be worn
    # in enclosed spaces outside private residences, July 30th
    #ax2.annotate("10",
    #             xy = (136, total[136]),
    #             xytext = (136, total[136] + arrow_adj),
    #             arrowprops = dict(facecolor = 'black', shrink = 0.05))
    # Labor Day, September 7th. Public schools to start class or are entering their second week
    #ax2.annotate("11",
    #            xy = (175, total[175]),
    #            xytext = (175, total[175] + arrow_adj),
    #            arrowprops = dict(facecolor = 'black', shrink = 0.05))
    ax2.set_xticks(range(len(daily_0_9)))
    ax2.set_xticklabels(dates, rotation = 300)
    # TO-DO: add xticks for every week
    for idx, lbl in enumerate(ax2.get_xticklabels()):
        if idx % 14 != 0:
            lbl.set_visible(False)
    plt.subplots_adjust(hspace = 0.5)
````
